```python
#importing libraries
import pandas as pd
import mysql.connector
import numpy as np
```

```python
# csv to dataframe
df_buses_1=pd.read_csv("C:/RedBus Project/Scrapped_Data/Assam.csv")
df_buses_2=pd.read_csv("C:/RedBus Project/Scrapped_Data/Andhra.csv")
df_buses_3=pd.read_csv("C:/RedBus Project/Scrapped_Data/kerala_data.csv")
df_buses_4=pd.read_csv("C:/RedBus Project/Scrapped_Data/Telangana.csv")
df_buses_5=pd.read_csv("C:/RedBus Project/Scrapped_Data/South_bengal.csv")
df_buses_6=pd.read_csv("C:/RedBus Project/Scrapped_Data/Himachal.csv")
df_buses_7=pd.read_csv("C:/RedBus Project/Scrapped_Data/Rajasthan.csv")
df_buses_8=pd.read_csv("C:/RedBus Project/Scrapped_Data/Uttarpradesh.csv")
df_buses_9=pd.read_csv("C:/RedBus Project/Scrapped_Data/Bihar.csv")
df_buses_10=pd.read_csv("C:/RedBus Project/Scrapped_Data/Westbengal.csv")

Final_df=pd.concat([df_buses_1,df_buses_2,df_buses_3,df_buses_4,df_buses_5,df_buses_6,
                    df_buses_7,df_buses_8,df_buses_9,df_buses_10],ignore_index=True)
Final_df
```

```python
Final_df.info()
```

```python
Final_df["Price"]=Final_df["Price"].str.replace("INR","")
Final_df["Price"]=Final_df["Price"].astype(float)
Final_df["Price"].fillna(0) 
```

```python
Final_df["Ratings"] = Final_df["Ratings"].astype(str)
```

```python
Final_df["Ratings"] = Final_df["Ratings"].str.split('\n').str[0]
```

```python
Final_df["Ratings"] = pd.to_numeric(Final_df["Ratings"], errors='coerce')
```

```python
Final_df["Ratings"] = Final_df["Ratings"].fillna(0)
```

```python
print(Final_df["Ratings"].tail())
```

```python
Final_df
```

```python
# info after the data type change
Final_df.info()
```

```python
Final_df = Final_df[Final_df["Price"] <= 7000]
```

```python
# replacing the nan value
Final_df = Final_df.replace({np.nan: None})
```

```python
# change dataframe to csv
path=r"C:/RedBus Project/Scrapped_Data/Final_busdetails_df.csv"
Final_df.to_csv(path,index=False)
```
